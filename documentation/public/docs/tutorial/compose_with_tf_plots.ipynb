```python
from graphviz import Source
```

```python
s = Source("""
digraph G {
  graph [rankdir=LR];
  node [shape=plaintext];
  a[label="input features"];
  b[label="neural network", shape=rect];
  d[label="decision forest", shape=rect];
  e[label="predictions"];
  a -> b;
  d -> e;
  b -> d [label="embedding"];
}""")
s.render("../image/tutorial/compose_with_tf_stack",format="png")
s
```

```python
s = Source("""
digraph G {
  graph [rankdir=LR];
  node [shape=plaintext];
  a [label="input features"];
  subgraph cluster_0 {
      label="ensemble";
      b [label="neural network", shape=rect];
      c [label="random forest", shape=rect];
      d [label="gradient boosted trees", shape=rect];
      e [label="average", shape=rect];
  }
  f [label="predictions"];
  a -> b -> e -> f;
  a -> c -> e;
  a -> d -> e;
}
""")
s.render("../image/tutorial/compose_with_tf_ensemble",format="png")
s
```
